```python
import numpy as np
import torch
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
X_train_tensor = torch.from_numpy(X_train).float()
X_test_tensor = torch.from_numpy(X_test).float()   
y_train_tensor = torch.from_numpy(y_train).float().view(-1, 1)
y_test_tensor = torch.from_numpy(y_test).float().view(-1, 1)
```

```python
X_train_tensor.shape
```

```python
import torch
import torch.nn as nn

class Model(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        
        # nn.Sequential - a container that stacks layers and runs them in order
        # data flows top to bottom automatically
        self.network = nn.Sequential(
            nn.Linear(num_features, 1),   # 30 inputs -> 1 output (one neuron)
            nn.Sigmoid()                   # squashes to 0-1 (probability)
        )

    def forward(self, features):
        out = self.network(features)   # one line - Sequential handles the chain
        return out
```

```python
learning_rate = 0.1
epochs = 25
```

```python
X_train_tensor.shape[1]
```

```python
model = Model(X_train_tensor.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
loss_function = nn.BCELoss()

for epoch in range(epochs):
    # 1. forward pass
    y_pred = model(X_train_tensor)
    
    # 2. loss
    loss = loss_function(y_pred, y_train_tensor)
    
    # 3. clear old gradients
    optimizer.zero_grad()        # replaces manual .grad.zero_()
    
    # 4. backward pass
    loss.backward()
    
    # 5. update weights
    optimizer.step()             # replaces manual w -= lr*grad
    
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
```

```python
for p in model.parameters():
    print(p)
```

```python
# see all parameters (weights + biases) from every layer
for name, param in model.named_parameters():
    print(name, param.shape)
```

```python
# training accuracy
with torch.no_grad():
    y_pred_train = model.forward(X_train_tensor)
    y_pred_train = (y_pred_train > 0.5).float().view(-1)
    train_acc = (y_pred_train == y_train_tensor.view(-1)).float().mean()
    print(f'Train Accuracy: {train_acc.item():.4f}')
```

```python
# test on UNSEEN data
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.5).float().view(-1)          
    test_acc = (y_pred == y_test_tensor.view(-1)).float().mean()
    print(f'Test Accuracy: {test_acc.item():.4f}')
```

## torch.optim is a PyTorch module that provides different optimizers.
## Optimizers update the model's weights and bias during training to reduce the loss.

## Common optimizers:
## SGD     -> Simple Gradient Descent optimizer.
## Adam    -> An adaptive optimizer (most commonly used).
## RMSprop -> Optimizer that adapts the learning rate during training.

## torch.optim also supports features like:
## - Learning rate scheduling (changing the learning rate during training)
## - Weight decay (L2 regularization) to help reduce overfitting.

## model.parameters() returns all trainable parameters of the model.
## These parameters include:
## - Weights: The weight matrices of layers like nn.Linear, nn.Conv2d, etc.
## - Biases: The bias values of those layers (if present).

## The optimizer uses these parameters and their gradients
## (computed during backpropagation) to update the model after each training step.

## model.train()   # put model in TRAINING mode (before training loop)
## model.eval()    # put model in EVAL mode (before testing/predicting)
