````python
import numpy as np
import torch
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
````

````python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
````

````python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
````

````python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

````python
X_train
````

````python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
````

````python
y_train
````

````python
X_train_tensor = torch.from_numpy(X_train).float()
X_test_tensor = torch.from_numpy(X_test).float()   
y_train_tensor = torch.from_numpy(y_train).float().view(-1, 1)
y_test_tensor = torch.from_numpy(y_test).float().view(-1, 1)
````

````python
X_train_tensor.shape
````

````python
from torch.utils.data import Dataset, DataLoader

class CustomDatset(Dataset):
    def __init__(self,features,labels):
        self.features=features
        self.labels=labels
    def __len__(self):
        return self.features.shape[0]
    def __getitem__(self,index):
        return self.features[index],self.labels[index]
````

````python
train_dataset=CustomDatset(X_train_tensor,y_train_tensor)
test_dataset=CustomDatset(X_test_tensor,y_test_tensor)
````

````python
len(train_dataset)
````

````python
len(test_dataset)
````

````python
train_dataset[10]
````

````python
train_loader=DataLoader(train_dataset,batch_size=32,shuffle=True)
test_loader=DataLoader(test_dataset,batch_size=32,shuffle=True)
````

````python
import torch
import torch.nn as nn   

# nn.Module = the base class for ALL PyTorch models
# it automatically gives you: weight tracking, gradient handling, GPU moving, saving/loading
class Model(nn.Module):

    def __init__(self, num_features):
        super().__init__()   # MUST call this - activates nn.Module's machinery (never skip)
        
        # nn.Sequential = a CONTAINER that stacks layers and runs them in order
        # data flows through top to bottom automatically
        # REPLACES writing each step manually in forward():
        #     out = self.linear(features)
        #     out = self.sigmoid(out)
        self.network = nn.Sequential(
            
            # nn.Linear(in, out) = a fully-connected layer
            # REPLACES the manual weight code:
            #     self.weights = torch.rand(X.shape[1], 1, requires_grad=True)
            #     self.bias = torch.zeros(1, requires_grad=True)
            #     z = torch.matmul(X, self.weights) + self.bias
            # nn.Linear creates the weights AND bias automatically, sets requires_grad=True,
            # does z = Wx + b, AND uses smart initialization (better than torch.rand)
            nn.Linear(num_features, 1),   # 30 inputs -> 1 output (one neuron)
            
            nn.Sigmoid()                   # squashes output to 0-1
        )

    def forward(self, features):
        out = self.network(features)   # ONE line - Sequential runs all layers in order
        return out
        
    # NOTE: no manual loss_function() anymore
    # we now use PyTorch's built-in nn.BCELoss()
````

````python
learning_rate = 0.1
epochs = 25
````

````python
X_train_tensor.shape[1]
````

````python
# create the model - nn.Linear auto-creates all 30 weights + 1 bias inside
model = Model(X_train_tensor.shape[1])

# ADAM OPTIMIZER
# REPLACES the manual weight update:
#     with torch.no_grad():
#         model.weights -= learning_rate * model.weights.grad
#         model.bias -= learning_rate * model.bias.grad
#
# model.parameters() = automatically grabs ALL weights & biases from every layer

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# BUILT-IN LOSS FUNCTION
# REPLACES the hand-written BCE formula:
#     loss = -(y * torch.log(y_pred) + (1-y) * torch.log(1-y_pred)).mean()
# nn.BCELoss() = the same Binary Cross-Entropy, pre-built and optimized
#                (it also handles the log(0) problem internally - no clamp needed!)
loss_function = nn.BCELoss()


for epoch in range(epochs):

    for batch_features, batch_labels in train_loader:
        # 1. FORWARD PASS - make predictions
        # use model(x), NOT model.forward(x) - model(x) also runs PyTorch's internal hooks
        y_pred = model(batch_features)
    
        # 2. LOSS
        loss = loss_function(y_pred, batch_labels.view(-1,1))
    
        # 3. CLEAR OLD GRADIENTS
        # REPLACES: model.weights.grad.zero_()  and  model.bias.grad.zero_()
        # optimizer.zero_grad() clears gradients of ALL parameters in one call
        # WHY? PyTorch ADDS UP gradients - without clearing, old ones pile up and corrupt training
        optimizer.zero_grad()
    
        # 4. BACKWARD PASS - autograd computes all gradients (the chain rule)
        loss.backward()
    
        # 5. UPDATE WEIGHTS
        # REPLACES the manual: w -= lr * w.grad
        # optimizer.step() applies Adam's update rule to every parameter automatically
        # (no need for torch.no_grad() - the optimizer handles that internally)
        optimizer.step()
    
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')
    # .item() = pulls the plain Python number out of a single-value tensor
````

````python
# model.parameters() = every weight and bias in the model
for p in model.parameters():
    print(p)
````

````python
# named_parameters() = same as parameters(), but with names (easier to read)
# with nn.Sequential, layers are numbered by their position:
#     index 0 = Linear, index 1 = Sigmoid
for name, param in model.named_parameters():
    print(name, param.shape)
    # -> network.0.weight torch.Size([1, 30])   (30 weights, one per feature)
    # -> network.0.bias torch.Size([1])          (1 bias)

# NOTE: with Sequential there's no "model.linear" anymore.
# to access a specific layer's weights, use the index: model.network[0].weight
````

````python
model.eval()   
accuracy_list = []

with torch.no_grad():   # no gradients needed for testing
    for batch_features, batch_labels in test_loader:
        y_pred = model(batch_features)
        y_pred = (y_pred > 0.5).float()   # threshold at 0.5
        # compare per batch, flatten both to match shapes
        batch_accuracy = (y_pred.view(-1) == batch_labels.view(-1)).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)   # average across batches
print(f'Accuracy: {overall_accuracy:.4f}')
````

## torch.optim is a PyTorch module that provides different optimizers.
## Optimizers update the model's weights and bias during training to reduce the loss.

## Common optimizers:
## SGD     -> Simple Gradient Descent optimizer.
## Adam    -> An adaptive optimizer (most commonly used).
## RMSprop -> Optimizer that adapts the learning rate during training.

## torch.optim also supports features like:
## - Learning rate scheduling (changing the learning rate during training)
## - Weight decay (L2 regularization) to help reduce overfitting.

## model.parameters() returns all trainable parameters of the model.
## These parameters include:
## - Weights: The weight matrices of layers like nn.Linear, nn.Conv2d, etc.
## - Biases: The bias values of those layers (if present).

## The optimizer uses these parameters and their gradients
## (computed during backpropagation) to update the model after each training step.

## model.train()   # put model in TRAINING mode (before training loop)
## model.eval()    # put model in EVAL mode (before testing/predicting)

## What nn.Module Replaced

| Manual code | Built-in replacement |
|-------------|---------------------|
| `torch.rand()` + `torch.zeros()` weights | `nn.Linear(in, out)` |
| `torch.matmul(X, w) + b` | `nn.Linear` (does it internally) |
| chaining steps in `forward()` | `nn.Sequential(...)` |
| hand-written BCE formula | `nn.BCELoss()` |
| `w -= lr * w.grad` | `optimizer.step()` |
| `w.grad.zero_()` | `optimizer.zero_grad()` |
| plain gradient descent | **Adam** (momentum + RMSprop) |



## `model.eval()` and `model.train()` — The Two Modes

A PyTorch model runs in one of two modes, and you switch between them depending on what you're doing.

### `model.train()` — Training Mode (the default)
Turns training behaviors **ON**:
- **Dropout** is active (randomly drops neurons)
- **BatchNorm** updates its running statistics from each batch

Use it before your training loop.

### `model.eval()` — Evaluation Mode
Turns those training behaviors **OFF**:
- **Dropout** is disabled (use ALL neurons for the real prediction)
- **BatchNorm** uses its stored running stats (not batch stats)

Use it before testing / making predictions.

---

### Why it matters

| | Training | Testing |
|---|----------|---------|
| Mode | `model.train()` | `model.eval()` |
| Dropout | ON (drops neurons) | OFF (all neurons) |
| BatchNorm | uses batch stats | uses saved running stats |

> If you **forget `model.eval()`** during testing, Dropout would randomly drop neurons while predicting → inconsistent, wrong results. And BatchNorm would use the test batch's stats instead of what it learned in training. Both hurt accuracy.

> **The pair works together:** `model.train()` before training, `model.eval()` before evaluating. Our model has no Dropout/BatchNorm yet, so it makes no difference *here* — but it's a critical habit for when you add them.

```python
model.train()   # before the training loop
# ... train ...

model.eval()    # before evaluation
with torch.no_grad():
    # ... predict ...
```
