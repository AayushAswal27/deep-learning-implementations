## Creating a tensor

```python
# torch.empty - reserves memory for the tensor but does NOT set the values
# so it contains whatever random numbers were already in that memory
x=torch.empty(3,3)
x
```

```python
# check type
type(x)
```

```python
# torch.zeros - filled entirely with 0s
torch.zeros(5,4)
```

```python
# torch.ones - filled entirely with 1s
torch.ones(3,4)
```

```python
# torch.rand - random values between 0 and 1
torch.rand(2,3)
```

## Reproducibility (seeds):

```python
# rand gives DIFFERENT numbers each run.
# manual_seed fixes the randomness so you get the SAME numbers every time
# (crucial for consistent, repeatable experiments)
torch.manual_seed(100)
torch.rand(3,3)
```

```python
torch.manual_seed(100)
torch.rand(3,3) # Same output because the seed is reset to 100
```

```python
#torch.tensor - build directly from a list of lists (your own data)
torch.tensor([[1,4,5,8],[9,4,2,71]])
```

```python
# arange(start, stop, step)
torch.arange(0,10,2)
```

```python
# linspace(start, stop, count) - 'count' evenly spaced numbers
torch.linspace(0,10,5)
```

```python
# eye(n) - identity matrix (1s on diagonal, 0s elsewhere)
torch.eye(4)
```

```python
# full(shape, value) - shape filled with a single value
torch.full((2,3),6)
```

## Tensor Shapes

```python
x=torch.tensor([[1,2,4,6],[5,6,7,4]])
x
```

```python
x.shape # 2 rows and 4 columns 
```

```python
# _like functions: make a new tensor with the SAME shape as x
print(torch.empty_like(x)) # empty, same shape
print(torch.zeros_like(x))   # zeros, same shape
print(torch.ones_like(x))     # ones, same shape
print(torch.rand_like(x, dtype=torch.float32))   # random; dtype needed (x is int)
```

## Tensor Data Types

```python
x.dtype #check the data type
```

```python
# assign dtype while creating a tensor 
print(torch.tensor([[1,2,5,9],[3,45,53,2]],dtype=torch.float64))
print(torch.tensor([[1.3,4.6,3.5],[2.9,4.1,9.6]], dtype=torch.int64))
```

```python
# .to(dtype) - CONVERT an existing tensor to a different type
x.to(torch.float64)
```

#  PyTorch Tensor Data Types (dtypes)

| Type | Meaning | When Used |
|------|---------|-----------|
| `torch.float32` | 32-bit float | **Default** for deep learning |
| `torch.float64` | 64-bit float | High-precision tasks |
| `torch.float16` | 16-bit float | Mixed-precision (saves GPU memory) |
| `torch.int64` | 64-bit integer | Indices (default int) |
| `torch.uint8` | 8-bit unsigned | Image pixels (0–255) |
| `torch.bool` | True/False | Masks for logic |
