## Reproducibility (seeds):

```python
# rand gives DIFFERENT numbers each run.
# manual_seed fixes the randomness so you get the SAME numbers every time
# (crucial for consistent, repeatable experiments)
torch.manual_seed(100)
torch.rand(3,3)
```

```python
torch.manual_seed(100)
torch.rand(3,3) # Same output because the seed is reset to 100
```

## Tensor Shapes

```python
x=torch.tensor([[1,2,4,6],[5,6,7,4]])
x
```

```python
x.shape # 2 rows and 4 columns 
```

```python
# _like functions: make a new tensor with the SAME shape as x
print(torch.empty_like(x)) # empty, same shape
print(torch.zeros_like(x))   # zeros, same shape
print(torch.ones_like(x))     # ones, same shape
print(torch.rand_like(x, dtype=torch.float32))   # random; dtype needed (x is int)
```

#  PyTorch Tensor Data Types (dtypes)

| Type | Meaning | When Used |
|------|---------|-----------|
| `torch.float32` | 32-bit float | **Default** for deep learning |
| `torch.float64` | 64-bit float | High-precision tasks |
| `torch.float16` | 16-bit float | Mixed-precision (saves GPU memory) |
| `torch.int64` | 64-bit integer | Indices (default int) |
| `torch.uint8` | 8-bit unsigned | Image pixels (0–255) |
| `torch.bool` | True/False | Masks for logic |

## Mathematical Operations

```python
#Scalar Operations — apply one number to every element
torch.manual_seed(100)
z=torch.rand(2,3)
print(z)
print(z+2)
print(z-2)
print(z*2)
print(z/3)
print(z%3)
print((z*100)//3)
```

```python
# Element-wise Operations — combine two SAME-shape tensors, position by position
torch.manual_seed(100)
a=torch.rand(2,3)
torch.manual_seed(101)
b=torch.rand(2,3)
print(a)
print(b)
print(a+b) # a[i]+b[i] at each position
print(a-b)
print(a*b) # element-wise multiply (NOT matrix multiply)
print(a/b)
```

```python
c = torch.tensor([1, -2, 3, -4]) 
torch.abs(c) # absolute value -> [1,2,3,4] (makes positive)
```

```python
torch.neg(c) # negate -> [-1,2,-3,4] (flips sign)
```

```python
d = torch.tensor([1.9, 2.2, 3.7, 4.4])
torch.round(d) # round to nearest -> [2,2,4,4]
```

```python
torch.ceil(d) # always round UP -> [2,3,4,5]
```

```python
torch.floor(d) # always round DOWN -> [1,2,3,4]
```

```python
d
```

```python
torch.clamp(d, min=2, max=3)  # Keep values between 2 and 3
```

```python
# torch.randint - random INTEGERS between low and high
# low = smallest possible value (included)
# high = upper limit (NOT included - stops one below)
torch.randint(low=0, high=10, size=(2,3))
```

```python
torch.manual_seed(100)
torch.randint(size=(2,3),low=5,high=10)
```


| Function | Produces | Range |
|----------|----------|-------|
| `torch.rand` | Random decimals (floats) | Between 0 and 1 |
| `torch.randint` | Random whole numbers (integers) | Between low and high |

```python
torch.manual_seed(101)
print(torch.rand(2,3))
torch.manual_seed(105)
print(torch.randint(size=(3,3),low=10,high=15))
```
