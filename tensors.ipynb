#  PyTorch Tensor Data Types (dtypes)

| Type | Meaning | When Used |
|------|---------|-----------|
| `torch.float32` | 32-bit float | **Default** for deep learning |
| `torch.float64` | 64-bit float | High-precision tasks |
| `torch.float16` | 16-bit float | Mixed-precision (saves GPU memory) |
| `torch.int64` | 64-bit integer | Indices (default int) |
| `torch.uint8` | 8-bit unsigned | Image pixels (0–255) |
| `torch.bool` | True/False | Masks for logic |

## Mathematical Operations

```python
#Scalar Operations — apply one number to every element
torch.manual_seed(100)
z=torch.rand(2,3)
print(z)
print(z+2)
print(z-2)
print(z*2)
print(z/3)
print(z%3)
print((z*100)//3)
```

```python
# Element-wise Operations — combine two SAME-shape tensors, position by position
torch.manual_seed(100)
a=torch.rand(2,3)
torch.manual_seed(101)
b=torch.rand(2,3)
print(a)
print(b)
print(a+b) # a[i]+b[i] at each position
print(a-b)
print(a*b) # element-wise multiply (NOT matrix multiply)
print(a/b)
```

```python
c = torch.tensor([1, -2, 3, -4]) 
torch.abs(c) # absolute value -> [1,2,3,4] (makes positive)
```

```python
torch.neg(c) # negate -> [-1,2,-3,4] (flips sign)
```

```python
d = torch.tensor([1.9, 2.2, 3.7, 4.4])
torch.round(d) # round to nearest -> [2,2,4,4]
```

```python
torch.ceil(d) # always round UP -> [2,3,4,5]
```

```python
torch.floor(d) # always round DOWN -> [1,2,3,4]
```

```python
d
```

```python
torch.clamp(d, min=2, max=3)  # Keep values between 2 and 3
```

```python
# torch.randint - random INTEGERS between low and high
# low = smallest possible value (included)
# high = upper limit (NOT included - stops one below)
torch.randint(low=0, high=10, size=(2,3))
```

```python
torch.manual_seed(100)
torch.randint(size=(2,3),low=5,high=10)
```


| Function | Produces | Range |
|----------|----------|-------|
| `torch.rand` | Random decimals (floats) | Between 0 and 1 |
| `torch.randint` | Random whole numbers (integers) | Between low and high |

```python
torch.manual_seed(101)
print(torch.rand(2,3))
torch.manual_seed(105)
print(torch.randint(size=(3,3),low=10,high=15))
```

## Reduction Operations — reduce many numbers into fewer

```python
torch.manual_seed(110)
l=torch.randint(size=(3,3),low=10,high=20,dtype=torch.float64)
l
```

```python
torch.sum(l) # add all elements
```

```python
torch.sum(l,dim=0) # sum DOWN each column
```

```python
torch.sum(l,dim=1) # sum ACROSS each row
```

```python
print(torch.mean(l))# average of all elements
print(torch.mean(l,dim=0)) # Average of each column
print(torch.mean(l,dim=1)) # Average of each row
```

```python
torch.median(l) # middle value
```

```python
print(torch.max(l)) # largest value
```

```python
torch.min(l) # smallest value
```

```python
torch.prod(l)  # multiply all elements together
```

```python
torch.std(l) # standard deviation (spread)
```

```python
torch.var(l)    # variance (spread squared)
```

```python
l
```

```python
# argmax/argmin return the INDEX (position), not the value
print(torch.argmax(l))         # position of the largest
print(torch.argmin(l))         # position of the smallest
```

## Matrix Operations 

```python
torch.manual_seed(120)
f = torch.randint(size=(2,3), low=0, high=10)
print(f)
torch.manual_seed(121)
g = torch.randint(size=(3,2), low=0, high=10)
print(g)
torch.matmul(f, g) # matrix multiplication
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])
torch.dot(vector1, vector2)    # dot product: 1*3 + 2*4 = 11
```

```python
f
```

```python
torch.transpose(f,0,1)       # swap dims 0 and 1 (rows <-> columns)
```

```python
torch.manual_seed(125)
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
print(h)
print(torch.det(h))        # determinant 
print(torch.inverse(h))    # inverse 
```

## Comparison Operations

```python
torch.manual_seed(85)
i = torch.randint(size=(2,3), low=0, high=10)
print(i)
torch.manual_seed(90)
j = torch.randint(size=(2,3), low=0, high=10)
print(j)

print(i > j)        
print(i < j)
print(i == j)         
print(i != j)         
print(i >= j)
print(i <= j)
```
