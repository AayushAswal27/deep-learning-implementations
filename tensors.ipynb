| `torch.float64` | 64-bit float | High-precision tasks |

## Mathematical Operations

```python
#Scalar Operations — apply one number to every element
torch.manual_seed(100)
z=torch.rand(2,3)
print(z)
print(z+2)
print(z-2)
print(z*2)
print(z/3)
print(z%3)
print((z*100)//3)
```

```python
# Element-wise Operations — combine two SAME-shape tensors, position by position
torch.manual_seed(100)
a=torch.rand(2,3)
torch.manual_seed(101)
b=torch.rand(2,3)
print(a)
print(b)
print(a+b) # a[i]+b[i] at each position
print(a-b)
print(a*b) # element-wise multiply (NOT matrix multiply)
print(a/b)
```

## In-place Operations

## Functions ending with (_)  modify the tensor directly (in place) instead of returning a new one — saves memory.

```python
torch.manual_seed(70)
m=torch.rand(2,3)
print(m)
torch.manual_seed(80)
n=torch.rand(2,3)
print(n)
```

```python
m+n # creates a new tensor 
```

```python
m.add_(n)  # m = m + n, stored back INTO m 
```

```python
m
```

```python
m.relu_() # apply relu directly to m
```

```python
m
```

## Copying a Tensor

```python
torch.manual_seed(71)
a=torch.rand(2,3)
print(a)
```

```python
b=a  #b = a does NOT copy — both names point to the same tensor.
b
```

```python
a[0][0]=12 # changes have been done in a variable here
a 
```

```python
b # but b also changes as both the variables are pointing to the same tensor 
```

```python
print(id(a),id(b)) # identical ids -> same object
```

```python
b=a.clone()
```

```python
a
```

```python
b
```

```python
a[0][0]=39 # changes have been done in 'a' variable
a
```

```python
b # b stays unchanged 
```

```python
print(id(a),id(b)) # different ids -> independent objects
```

## b = a shares memory. Use .clone() when you need a separate copy.

## Reshaping

```python
#rearranges the numbers into a new shape 
x = torch.arange(12)  # tensor([0,1,2,...,11]) - 12 numbers in a line
print(x)
print(x.shape)
print(x.reshape(3, 4))        # arrange into 3 rows x 4 cols
print(x.reshape(4, 3))        # or 4 rows x 3 cols
print(x.reshape(2, 6))        # or 2 rows x 6 cols
print(x.reshape(2,2,3))
```

```python
# you can use -1 to auto calculate 
x.reshape(3, -1)   # we have given rows=3, PyTorch figures out cols = 4 automatically
```

```python
#view() — same as reshape
x = torch.arange(12)
print(x.view(3, 4))    # same result as reshape(3,4)
print(x.view(2, 6))
```

##  reshape() and view() both reshape tensors.
##  reshape() is safer, view() is faster.

## squeeze() — remove dimensions of size 1

```python
torch.manual_seed(30)
w=torch.rand(2,3,1)
print(w)
print(w.shape)
```

```python
w.squeeze().shape
```

```python
# we can also squeeze the specific dimension also
w.squeeze(0).shape  # Remove dim 0 only if its size is 1
# Dim 0 is not removed because its size is 2.
```

```python
w.squeeze(2).shape
```

```python
# unsqueeze - add a dimension of size 1
x = torch.rand(226,226,3)            
print(x.unsqueeze(0).shape)  
print(x.unsqueeze(3).shape)   
```

## Why you NEED this: neural networks expect a batch dimension. A single image of shape (3, 224, 224) must become (1, 3, 224, 224) — "a batch of 1 image" — before feeding it to a model. unsqueeze(0) adds that batch dimension

```python
## permute() — reorder the dimensions
img = torch.rand(2, 3, 4)         # shape (2, 3, 4)
print(img.permute(2, 0, 1).shape) # torch.Size([4, 2, 3])
```

## Why you NEED this — images!  Image libraries store images as (Height, Width, Channels) = HWC. But PyTorch wants (Channels, Height, Width) = CHW.

```python
 #flatten() — collapse to 1D
# flatten - collapse to 1D
t = torch.rand(2, 3, 4)  # shape (2, 3, 4) = 24 numbers
print(t.flatten().shape)  # torch.Size([24]) - all in one line
```

```python
t
```

```python
t.flatten(start_dim=1)   # keeps dim 0 (batch), flattens the rest
# shape (2, 3, 4) → (2, 12)   
```

## Joining & Indexing


```python
a = torch.tensor([1, 2])
b = torch.tensor([3, 4])
```

```python
# Combine tensors into one
torch.stack([a, b])  # [[1,2],[3,4]] shape (2,2)
```

```python
# cat - join along an EXISTING dimension (makes a longer tensor)
torch.cat([a, b])    # [1,2,3,4] (no new dimension)
```

```python
# indexing & slicing (same syntax as NumPy/Python lists)
t = torch.tensor([[10,20,30],
                  [40,50,60]])
print(t[0])# first row -> [10,20,30]
print(t[0][1])       # row0, col1 -> 20
print(t[:, 0])       # all rows, col 0 -> [10,40]
print(t[1, :])       # row 1, all cols -> [40,50,60]
print(t[:, 1:3])     # all rows, cols 1-2 -> [[20,30],[50,60]]
```
